```python
import numpy as np
import pandas as pd 
import sys
import io 
```

```python
from google.colab import files 
uploaded = files.upload()
```

```python
age = pd.read_csv(io.BytesIO(uploaded['age2.csv']))
race = pd.read_csv(io.BytesIO(uploaded['race2.csv']))
sex =  pd.read_csv(io.BytesIO(uploaded['sex2.csv']))
```

```python
result = pd.concat([age,race,sex], axis=1)
result
```

```python
result.to_csv('result.csv') 
files.download('result.csv')
```
