```python
import numpy as np
import pandas as pd 
import sys
import io 
```

```python
from google.colab import files 
uploaded = files.upload()
```

```python
death = pd.read_csv(io.BytesIO(uploaded['result_death3.csv']))
comfirm = pd.read_csv(io.BytesIO(uploaded['result_comfirm3.csv']))
```

```python
comfirm
```

```python
output = pd.merge(comfirm,death,how='outer')
```

```python
output=output.fillna(0)
```

```python
output["Newly Reported Death Count"]=output["Newly Reported Death Count"].astype(int)
output["Cumulative Death Count"]=output["Cumulative Death Count"].astype(int)
output["Cumulative Confirmed Case Count"]=output["Cumulative Confirmed Case Count"].astype(int)
output["Newly Confirmed Case Count"]=output["Newly Confirmed Case Count"].astype(int)
```

```python
new = output
new
```

```python
new[new["Newly Confirmed Case Count"]<0]

new[new["Newly Reported Death Count"]<0]
```

```python
new.loc[new["Newly Confirmed Case Count"] < 0, "Newly Confirmed Case Count"] = 0
new.loc[new["Newly Reported Death Count"] < 0, "Newly Reported Death Count"] = 0
```

```python
new
```

```python
new.to_csv('new.csv') 
files.download('new.csv')
```
