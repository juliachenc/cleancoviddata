```python
from google.colab import files 
uploaded = files.upload()
```

```python
import numpy as np
import pandas as pd 
import sys
import io 
```

```python
df1 = pd.read_csv(io.BytesIO(uploaded['df1.csv']))
df1 = df1.fillna(0)

df2 = pd.read_csv(io.BytesIO(uploaded['df2.csv']))
df2 = df2.fillna(0)

df3 = pd.read_csv(io.BytesIO(uploaded['df3.csv']))
df3 = df3.fillna(0)
```

```python
df1['City/Community'] = 'City of ' + df1['City/Community'].astype(str)
df1 = df1.rename(columns={'City/Community': 'Community'})
df1
```

```python
df2['Neighborhood in Los Angeles City'] = 'Los Angeles - ' + df2['Neighborhood in Los Angeles City'].astype(str)
df2 = df2.rename(columns={'Neighborhood in Los Angeles City': 'Community'})
df2
```

```python
df3['Unincorporated Area in LA County'] = 'Unincorporated - ' + df3['Unincorporated Area in LA County'].astype(str)
df3 = df3.rename(columns={'Unincorporated Area in LA County': 'Community'})
df3
```

```python
frames = [df1, df2, df3]
result = pd.concat(frames,ignore_index=True,verify_integrity=True)
result
```

```python
new1= result.groupby(["Data Type","Community"]).sum()
new1.head(4)
```

```python
new2= new1.unstack().transpose()
new2
```

```python
population_table = new2[0:342]
long_data = new2[342::]
population_table
```
